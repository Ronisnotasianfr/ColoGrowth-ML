# 01. Exploratory Data Analysis (EDA)

In this notebook, we load, inspect, and analyze our datasets: GSE39582 (microarray data) and TCGA-COAD (RNA-seq data).

### Objectives:
1. Overview of data size, types, and missing values.
2. Visualize distribution of gene expression levels.
3. Investigate clinical characteristics (age, stage, tumor size, gender).
4. Assess sample demographics and prepare for preprocessing.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

print("Libraries imported successfully.")
```

## Load/Generate Datasets

```python
import sys
sys.path.append('..')
from src.preprocess import generate_synthetic_data

# Let's generate synthetic data for analysis
expr_df, clinical_df = generate_synthetic_data(n_samples=250, n_genes=1200)
print(f"Expression matrix shape: {expr_df.shape}")
print(f"Clinical metadata shape: {clinical_df.shape}")
```

## 1. Inspect Clinical Metadata

```python
print("Clinical Metadata Columns:")
print(clinical_df.info())
print("\nFirst few samples:")
display(clinical_df.head())
```

### Visualize Clinical Distributions

We check the age distribution, gender balance, and stage distributions.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Age Distribution
sns.histplot(data=clinical_df, x='age', kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Age Distribution')

# Tumor Size
sns.histplot(data=clinical_df, x='tumor_size_mm', kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Tumor Size (mm) Distribution')

# Gender distribution
sns.countplot(data=clinical_df, x='gender', ax=axes[1, 0], palette='pastel')
axes[1, 0].set_title('Gender Distribution')

# Stage distribution
sns.countplot(data=clinical_df, x='stage', ax=axes[1, 1], palette='Set2',
              order=['Stage I', 'Stage II', 'Stage III', 'Stage IV'])
axes[1, 1].set_title('TNM Stage Distribution')

plt.tight_layout()
plt.show()
```

## 2. Analyze Gene Expression Data

```python
print(f"Expression values range: [{expr_df.min().min():.2f}, {expr_df.max().max():.2f}]")

# Plot distribution of sample means and variances
sample_means = expr_df.mean(axis=1)
sample_vars = expr_df.var(axis=1)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(sample_means, kde=True, ax=axes[0], color='purple')
axes[0].set_title('Distribution of Sample Mean Expressions')
axes[0].set_xlabel('Mean Expression')

sns.histplot(sample_vars, kde=True, ax=axes[1], color='teal')
axes[1].set_title('Distribution of Sample Expression Variances')
axes[1].set_xlabel('Variance')
plt.show()
```

### Top Highly Variable Genes

```python
gene_vars = expr_df.var(axis=0)
top_20_var = gene_vars.nlargest(20)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_20_var.values, y=top_20_var.index, palette='viridis')
plt.title('Top 20 Most Variable Genes')
plt.xlabel('Variance')
plt.show()
```
